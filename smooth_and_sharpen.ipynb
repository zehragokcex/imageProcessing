```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def smooth_and_sharpen(image, smoothing_kernel_size, sharpening_factor):
    # Smoothing (Düzleştirme)
    smoothed_img = cv2.GaussianBlur(image, (smoothing_kernel_size, smoothing_kernel_size), 0)

    # Sharpening (Keskinleştirme)
    sharpening_kernel = np.array([[-1, -1, -1],
                                  [-1,  9, -1],
                                  [-1, -1, -1]])
    sharpened_img = cv2.filter2D(smoothed_img, -1, sharpening_factor * sharpening_kernel)

    return smoothed_img, sharpened_img

# Resmi oku
img = cv2.imread('bit.png', 0)  # 0, gri tonlamalı okuma

# Smoothing ve Sharpening uygula
smoothing_kernel_size = 5  # Düzleştirme kernel boyutunu belirle (örneğin, 5x5)
sharpening_factor = 1.5   # Keskinleştirme faktörünü belirle

smoothed_image, sharpened_image = smooth_and_sharpen(img, smoothing_kernel_size, sharpening_factor)

# Orjinal, Düzleştirilmiş ve Keskinleştirilmiş resimleri görselleştirme
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
plt.title('Orjinal Resim')
plt.imshow(img, cmap='gray')
plt.axis('off')

plt.subplot(1, 3, 2)
plt.title('Düzleştirilmiş Resim')
plt.imshow(smoothed_image, cmap='gray')
plt.axis('off')

plt.subplot(1, 3, 3)
plt.title('Keskinleştirilmiş Resim')
plt.imshow(sharpened_image, cmap='gray')
plt.axis('off')

plt.show()
```
