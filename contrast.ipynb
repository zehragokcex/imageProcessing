```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def contrast_stretching(image, min_out=0, max_out=255):
    
    normalized_img = cv2.normalize(image, None, alpha=0, beta=1, norm_type=cv2.NORM_MINMAX, dtype=cv2.CV_32F)
    
    
    stretched_img = np.clip((normalized_img - normalized_img.min()) / (normalized_img.max() - normalized_img.min()) * (max_out - min_out) + min_out, 0, 255)

    
    stretched_img = stretched_img.astype(np.uint8)

    return stretched_img

# Resmi oku
img = cv2.imread('bit.png', 0)  # 0, gri tonlamal okuma

# Kontrast stretching uygula
stretched_img = contrast_stretching(img)

# Orjinal ve kontrast gerilmi resmi grselletirme
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.title('Orjinal Resim')
plt.imshow(img, cmap='gray')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.title('Kontrast Stretching')
plt.imshow(stretched_img, cmap='gray')
plt.axis('off')

plt.show()
```
