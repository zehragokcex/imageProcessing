```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def average_filter(image, kernel_size):
    # Kernel olutur
    kernel = np.ones((kernel_size, kernel_size), np.float32) / (kernel_size * kernel_size)

    # Filtre uygula
    filtered_img = cv2.filter2D(image, -1, kernel)

    return filtered_img

# Resmi oku
img = cv2.imread('bit.png', 0)  # 0, gri tonlamal okuma

# Ortalama filtresi uygula
kernel_size = 5  # Filtre boyutunu belirle (rnein, 5x5)
smoothed_img = average_filter(img, kernel_size)

# Orjinal ve ortalama filtre uygulanm resmi grselletirme
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.title('Orjinal Resim')
plt.imshow(img, cmap='gray')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.title('Ortalama Filtre Uygulanm')
plt.imshow(smoothed_img, cmap='gray')
plt.axis('off')

plt.show()
```
