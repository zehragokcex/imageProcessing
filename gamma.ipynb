```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def gamma_correction(image, gamma):
    # Normalize pixel values to the range [0, 1]
    normalized_img = image / 255.0
    
    # Apply gamma correction
    gamma_corrected_img = np.power(normalized_img, gamma)
    
    # Scale back to the range [0, 255]
    gamma_corrected_img = (gamma_corrected_img * 255).astype(np.uint8)

    return gamma_corrected_img

# Resmi oku
img = cv2.imread('histogram.png', 0)  # 0, gri tonlamal okuma

# Gamma dzeltme uygula
gamma_value = 1.5  # Gamma deeri, 1'den byk bir say seebilirsiniz
gamma_corrected_img = gamma_correction(img, gamma_value)

# Orjinal ve gamma dzeltilmi resmi grselletirme
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.title('Orjinal Resim')
plt.imshow(img, cmap='gray')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.title('Gamma Dzeltme (Gamma = {})'.format(gamma_value))
plt.imshow(gamma_corrected_img, cmap='gray')
plt.axis('off')

plt.show()
```
