```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Resmi oku
img = cv2.imread('histogram.png', 0)  # 0, gri tonlamalı okuma


if img is None:
    print("Error: Unable to load the image.")
else:
    
    print("Image Type:", type(img))
    print("Image Shape:", img.shape)

    # Histogram eşitleme işlemi
    equ = cv2.equalizeHist(img)

    # Orjinal ve eşitlenmiş resmi görselleştirme
    plt.figure(figsize=(12, 6))

    plt.subplot(1, 3, 1)
    plt.title('Orjinal Resim')
    plt.imshow(img, cmap='gray')
    plt.axis('off')

    plt.subplot(1, 3, 2)
    plt.title('Histogram Eşitleme')
    plt.imshow(equ, cmap='gray')
    plt.axis('off')

    plt.subplot(1, 3, 3)
    plt.title('Histogramlar')
    plt.hist(img.flatten(), 256, [0, 256], color='r', alpha=0.5, label='Orjinal')
    plt.hist(equ.flatten(), 256, [0, 256], color='b', alpha=0.5, label='Eşitlenmiş')
    plt.legend(loc='upper right')
    plt.show()

```
