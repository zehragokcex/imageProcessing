```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def opening_closing_demo(image_path, operation_type, kernel_size):
    # Resmi oku
    img = cv2.imread('bit.png', 0)  # Gri tonlamal okuma

    # Morfolojik ilem tipini belirle
    if operation_type == 'opening':
        operation = cv2.MORPH_OPEN
    elif operation_type == 'closing':
        operation = cv2.MORPH_CLOSE
    else:
        raise ValueError("Geersiz morfolojik ilem tipi. 'opening' veya 'closing' kullann.")

    # Morfolojik ilem iin kernel olutur
    kernel = np.ones((kernel_size, kernel_size), np.uint8)

    # Morfolojik ilem uygula
    result_img = cv2.morphologyEx(img, operation, kernel)

    # Orjinal ve ilem sonras resimleri grselletirme
    plt.figure(figsize=(12, 6))

    plt.subplot(1, 2, 1)
    plt.title('Orjinal Resim')
    plt.imshow(img, cmap='gray')
    plt.axis('off')

    plt.subplot(1, 2, 2)
    plt.title(f'{operation_type.capitalize()} lemi (Kernel Boyutu: {kernel_size})')
    plt.imshow(result_img, cmap='gray')
    plt.axis('off')

    plt.show()

# rnek kullanm
opening_closing_demo('ornek_resim.jpg', 'opening', 5)
```
