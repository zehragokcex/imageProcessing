```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def rgb_thresholding(image, threshold):
    # Eşik değeri üzerindeki pikselleri beyaz, altındakileri siyah yap
    thresholded_image = np.zeros_like(image)
    thresholded_image[image > threshold] = 255

    return thresholded_image

# Resmi oku
img = cv2.imread('bit.png')

# Resmi RGB renk uzayına dönüştür
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Eşik değerini belirle
threshold_value = 120

# RGB kullanarak bölütlenmiş resmi al
thresholded_img = rgb_thresholding(img_rgb, threshold_value)

# Orjinal ve bölütlenmiş resmi görselleştirme
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.title('Orjinal Resim')
plt.imshow(img_rgb)
plt.axis('off')

plt.subplot(1, 2, 2)
plt.title('RGB Bölütlenmiş Resim')
plt.imshow(thresholded_img, cmap='gray')
plt.axis('off')

plt.show()
```
