```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def gaussian_filter(image, kernel_size, sigma):
    # Gaussian kernel oluştur
    kernel = cv2.getGaussianKernel(kernel_size, sigma)
    kernel_2d = np.outer(kernel, kernel.T)

    # Filtre uygula
    filtered_img = cv2.filter2D(image, -1, kernel_2d)

    return filtered_img

# Resmi oku
img = cv2.imread('bit.png', 0)  # 0, gri tonlamalı okuma

# Gaussian filtresi uygula
kernel_size = 5  # Filtre boyutunu belirle (örneğin, 5x5)
sigma = 1.0  # Gaussian filtre parametresi
smoothed_img = gaussian_filter(img, kernel_size, sigma)

# Orjinal ve Gaussian filtresi uygulanmış resmi görselleştirme
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.title('Orjinal Resim')
plt.imshow(img, cmap='gray')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.title('Gaussian Filtre Uygulanmış')
plt.imshow(smoothed_img, cmap='gray')
plt.axis('off')

plt.show()
```
