```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def add_salt_and_pepper_noise(image, salt_prob, pepper_prob):
    noisy_image = np.copy(image)

    # Salt noise ekle
    salt_noise = np.random.rand(*image.shape) < salt_prob
    noisy_image[salt_noise] = 255

    # Pepper noise ekle
    pepper_noise = np.random.rand(*image.shape) < pepper_prob
    noisy_image[pepper_noise] = 0

    return noisy_image

# Resmi oku
img = cv2.imread('bit.png', 0)  # 0, gri tonlamalı okuma

# Salt-and-pepper gürültüsü ekle
salt_prob = 0.02  # Salt gürültü olasılığı
pepper_prob = 0.02  # Pepper gürültü olasılığı
noisy_img = add_salt_and_pepper_noise(img, salt_prob, pepper_prob)

# Orjinal ve salt-and-pepper gürültülü resmi görselleştirme
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.title('Orjinal Resim')
plt.imshow(img, cmap='gray')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.title('Salt-and-Pepper Gürültülü Resim')
plt.imshow(noisy_img, cmap='gray')
plt.axis('off')

plt.show()
```
