```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def laplace_blur(image, kernel_size):
    # Laplace kernel oluştur
    laplace_kernel = np.array([[0, 1, 0],
                               [1, -4, 1],
                               [0, 1, 0]])

    # Filtre uygula
    laplace_filtered_img = cv2.filter2D(image, cv2.CV_64F, laplace_kernel)

    # Görüntüyü normalize et
    laplace_filtered_img = cv2.normalize(laplace_filtered_img, None, alpha=0, beta=255, norm_type=cv2.NORM_MINMAX, dtype=cv2.CV_8U)

    return laplace_filtered_img

# Resmi oku
img = cv2.imread('bit.png', 0)  # 0, gri tonlamalı okuma

# Laplace blurring uygula
kernel_size = 3  # Filtre boyutunu belirle (örneğin, 3x3)
laplace_blurred_img = laplace_blur(img, kernel_size)

# Orjinal ve Laplace blurring uygulanmış resmi görselleştirme
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.title('Orjinal Resim')
plt.imshow(img, cmap='gray')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.title('Laplace Blurring')
plt.imshow(laplace_blurred_img, cmap='gray')
plt.axis('off')

plt.show()
```
