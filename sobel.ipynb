```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def sobel_filter(image):
    # Gradyanları hesapla
    grad_x = cv2.Sobel(image, cv2.CV_64F, 1, 0, ksize=3)
    grad_y = cv2.Sobel(image, cv2.CV_64F, 0, 1, ksize=3)

    # Gradyan büyüklüğünü ve açısını hesapla
    grad_magnitude = np.sqrt(grad_x**2 + grad_y**2)
    grad_angle = np.arctan2(grad_y, grad_x)

    return grad_magnitude, grad_angle

# Resmi oku
img = cv2.imread('bit.png', 0)  # 0, gri tonlamalı okuma

# Sobel filtresi uygula
gradient_magnitude, gradient_angle = sobel_filter(img)

# Orjinal ve Sobel filtresi uygulanmış görüntüyü görselleştirme
plt.figure(figsize=(12, 6))

plt.subplot(1, 3, 1)
plt.title('Orjinal Resim')
plt.imshow(img, cmap='gray')
plt.axis('off')

plt.subplot(1, 3, 2)
plt.title('Sobel Gradyan Büyüklüğü')
plt.imshow(gradient_magnitude, cmap='gray')
plt.axis('off')

plt.subplot(1, 3, 3)
plt.title('Sobel Gradyan Açısı')
plt.imshow(gradient_angle, cmap='hsv')
plt.axis('off')

plt.show()
```
